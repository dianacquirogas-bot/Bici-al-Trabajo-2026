## **PREGUNTA 4**
En días laborales, ¿la media de cnt ≥ 4.500? (plantea y resuelve un test de hipótesis de una muestra).

```python
# 1. Filtrar datos de días laborales
cnt_laborales = df[df['workingday'] == 1]['cnt']

# Valor de referencia para la prueba
valor_hipotesis = 4500

# Realizar el test de hipótesis de una muestra (t-test)
# H0: media <= 4500
# Ha: media > 4500 (prueba unilateral derecha)

# Como la muestra es grande (n > 30) y queremos probar la media,
# podemos usar un t-test. Scipy's ttest_1samp realiza una prueba bilateral por defecto.
# Para una prueba unilateral derecha, podemos usar el p-valor bilateral y dividirlo por 2,
# y comparar con alpha. O, más rigurosamente, ajustar el p-valor si el estadístico t es positivo.

t_statistic, p_value = stats.ttest_1samp(cnt_laborales, valor_hipotesis)

# Para una prueba unilateral derecha (Ha: media > valor_hipotesis):
# Si el estadístico t es positivo, el p-valor unilateral es p_value / 2.
# Si el estadístico t es negativo, el p-valor unilateral es 1 - (p_value / 2),
# o más simply, dado que estamos probando si la media es MAYOR, un t negativo
# ya nos dice que no hay evidencia para rechazar H0 en favor de Ha.
# Una forma más sencilla es simplemente verificar si t > 0 y p_value / 2 < alpha.
# Regla de decision

alpha = 0.05 # Nivel de significancia

print(f"Estadístico t: {t_statistic}")
print(f"P-valor (bilateral): {p_value}")

# Interpretación para prueba unilateral derecha
if t_statistic > 0 and p_value / 2 < alpha:
    print(f"\nCon un nivel de significancia de {alpha}, rechazamos la hipótesis nula porque tenemos evidencia estadistica de que la media de la de,amda es significativamente diferente a 4500.")
    print("Hay evidencia suficiente para concluir que la media de cnt en días laborales es mayor a 4500.")
else:
    print(f"\nCon un nivel de significancia de {alpha}, no rechazamos la hipótesis nula.")
    print("No hay evidencia suficiente para concluir que la media de cnt en días laborales es mayor a 4500.")

# También podemos calcular el intervalo de confianza para la media si es relevante
# (aunque la pregunta se centra en el test de hipótesis)
# mean = cnt_laborales.mean()
# se = stats.sem(cnt_laborales) # Standard error of the mean
# confidence_interval = stats.t.interval(1 - alpha, len(cnt_laborales)-1, loc=mean, scale=se)
# print(f"Intervalo de confianza del {((1 - alpha)*100):.0f}% para la media: {confidence_interval}")
```

## **PREGUNTA 5**
¿Cuál es un IC ( intervalo de confianza) bootstrap para la mediana de cnt y cómo cambia tu mensaje si la distribución es sesgada?

```python
# Número de remuestreos
n_iterations = 5000
sample_size = len(df['cnt'])
medians = []

# remuestreo
for _ in range(n_iterations):
    bootstrap_sample = df['cnt'].sample(n=sample_size, replace=True)
    bootstrap_median = bootstrap_sample.median()
    medians.append(bootstrap_median)
# intervalo de confianza
intervalo_confianza = np.percentile(medians, [2.5, 97.5])

print(f"IC de la mediana: {intervalo_confianza}")

```
